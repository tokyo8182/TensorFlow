```python
from statsmodels.tsa.stattools import adfuller


# Convert 'date' column to datetime type and set it as the DataFrame index
#log_coal_df['year'] = pd.to_datetime(log_coal_df['year'])
#log_coal_df.set_index('year', inplace=True)

# Function to check stationarity using ADF test
def check_stationarity(series):
    result = adfuller(series, autolag='AIC')
    p_value = result[1]
    return p_value

# Function to apply differencing and check stationarity
def find_stationary_series(series, max_iterations=4):
    p_value = check_stationarity(series)
    iteration = 0

    while p_value >= 0.05 and iteration < max_iterations:
        # Apply first-order differencing to the series
        differenced_series = series.diff().dropna()
        p_value = check_stationarity(differenced_series)
        series = differenced_series
        print('iteration : ', iteration+1)
        print('p_value :', p_value)
        iteration += 1
        print('*'*50)


    return series

```

# **Reversing Differencing and Log Transformation on Co2**

```python
import numpy as np

# Step 1: Undo differencing
def inverse_difference(series, initial_value, n_diff):
    inverted = []
    inverted.append(initial_value)
    for i in range(n_diff, len(series)):
        inverted_value = inverted[i - n_diff] + series[i]
        inverted.append(inverted_value)
    return inverted

# Step 2: Inverse log transformation
forecast_data= np.exp(inverse_difference(forecast, log_co2_df['Log_Transformed'].iloc[-1], 1))
```

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
import numpy as np



# Define a list of possible values for p, d, and q
p_values = range(0, 30)  # You can adjust the range based on your requirements
d_values = range(0, 10)  # You can adjust the range based on your requirements
q_values = range(0, 5)  # You can adjust the range based on your requirements

best_mae = np.inf
best_order = None

# Try different combinations of p, d, and q
for p in p_values:
    for d in d_values:
        for q in q_values:
            try:
                # Fit ARIMA model with the current parameters
                model = ARIMA(co2_train, order=(p, d, q))
                model_fit = model.fit()

                # Make predictions on the validation set
                forecast = model_fit.forecast(steps=len(co2_test))

                # Calculate Mean Absolute Error (MAE) as the evaluation metric
                mae = mean_absolute_error(co2_test, forecast)

                # Update the best model if the current model has lower MAE
                if mae < best_mae:
                    best_mae = mae
                    best_order = (p, d, q)

            except Exception as e:
                # Skip the combination if the model fitting fails
                continue

# Fit the best ARIMA model with the selected order parameters
best_model = ARIMA(co2_train, order=(best_order[0], best_order[1], best_order[2]))
best_model_fit = best_model.fit()

# Print the summary of the best ARIMA model
print(best_model_fit.summary())

# Make predictions on the test set using the best ARIMA model
forecast3 = best_model_fit.forecast(steps=10)

# Undo differencing and log transformation to get the forecasted values in the original scale
forecast_original3 = np.exp(inverse_difference(forecast3, log_co2_df['Log_Transformed'].iloc[-1], 1))
```
